```python
from __future__ import print_function
import nilmtk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = nilmtk.DataSet('../data/UKDALE/UKDALE.h5')
```

```python
dataset.set_window("2014-06-01", "2014-06-02")
```

```python
BUILDING = 1
elec = dataset.buildings[BUILDING].elec
fridge = elec['fridge']
wm = elec['washing machine']
```

```python
activations_f = fridge.get_activations()
activations_wm = wm.get_activations()
```

```python
print("Number of activations =", len(activations_f))
print("Number of activations =", len(activations_wm))
```

```python
for activation in activations_wm:
    activation.plot()

for activation in activations_f:
    activation.plot()



plt.show()
```
